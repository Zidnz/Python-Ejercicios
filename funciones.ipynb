# Funciones


## Una función entre conjuntos $A$ y $B$, denotada por $f: A → B$ son parejas ordenadas $(a,b)$ de tal manera que a cada elemento del primer conjunto A le corresponde un solo elemento del segundo conjunto B.


## Al conjunto $A$ le llamamos dominio de la función y al conjunto $B$ le llamamos codominio, imagen o rango de la función.

```python
import pylab as plt
from matplotlib_venn import venn2, venn3, venn3_circles

v = venn2(subsets=(1,1,0,1,0))
v.get_label_by_id('100').set_text('Dominio')
v.get_label_by_id('010').set_text('Imagen')
plt.title("Funciones")
plt.show()
```

## De la definición tenemos que si $(a,b)$ y $(a,c)$ son parejas ordenadas de la función $f$ entonces $b = c$.

# ¿Cómo gráficar?

### Primero identificamos la función a graficar, en general tienen la siguiente estructura: $f(x)=x\;$,$y=x^2+2x+1$

### Después determinamos su dominio e imagen, teniendo en cuenta si estos tienen restricciones. En general las funciones que tienen restricciones estan escritas de forma fraccionarias donde el denominador tiene puntos donde se anula como $\frac{1}{x}$ en $x = 0$.

```python
from sympy.interactive.printing import init_printing
init_printing(use_unicode=False, wrap_line=False)
from sympy.matrices import Matrix, eye, zeros, ones, diag, GramSchmidt
M = Matrix([[-3,"f(-3)"], [-2,"f(-2)"], [-1,"f(-1)"], [0,"f(0)"], [1,"f(1)"], [2,"f(2)"], [3,"f(3)"]]); M
```

### Es decir, damos distintos valores de $x$ para la función $f(x)$ y los guardamos en una tabla, se recomienda que sean numeros enteros cercanos al 0, con lo que obtenemos las parejas $(x,f(x))$ que marcamos en el plano cartesiano.

```python
# importar todas las funciones de pylab
from pylab import *

# importar el módulo pyplot
import matplotlib.pyplot as plt
```

```python
from math import *
from numpy import *

t = arange(-3, 5, 3)

y1 = t
y2 = t*2
p1, p2 = plot(t, y1, t, y2)

# Texto en la gráfica en coordenadas (x,y)
texto1 = text(-3.1,-1.7, r'$f\,(x)=x$', fontsize=16)
texto2 = text(0.8,5, r'$f\,(x)=2x$', fontsize=16)

#Tabulación de f(x) = x
y1marca1 = text(2,2, r'$\bullet$', fontsize=20)
y1marca2 = text(1,1, r'$\bullet$', fontsize=20)
y1marca3 = text(0,0, r'$\bullet$', fontsize=20)
y1marca4 = text(-1,-1, r'$\bullet$', fontsize=20)
y1marca5 = text(-2,-2, r'$\bullet$', fontsize=20)
y1marca6 = text(-3,-3, r'$\bullet$', fontsize=20)

#Tabulación de f(x) = 2x
y1marca1 = text(2,4, r'$\bullet$', fontsize=20)
y1marca2 = text(1,2, r'$\bullet$', fontsize=20)
y1marca4 = text(-1,-2, r'$\bullet$', fontsize=20)
y1marca5 = text(-2,-4, r'$\bullet$', fontsize=20)
y1marca6 = text(-3,-6, r'$\bullet$', fontsize=20)


# Añado una malla al gráfico
grid()

title('Representacion de dos gráficas')
xlabel('Valores de x')
ylabel('Valores de f(x)')

show()
```
